```python
import pandas as pd 
import jieba
import gensim
```

```python
df = pd.read_csv("td.csv")
```

```python
strs =[]
for i in df['comment']:
    strs.append(list(jieba.cut(i)))
    
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
#轉成doc2vec所需的格式 
#ex: [['你好','嗎'],['預測','檔案']] ===> [TaggedDocments(['你好','嗎'], ['0']),TaggedDocments(['預測','檔案'],['1'])]

def to_tagged(docs):
    re = []
    j = 0
    for i in docs:
        re.append(TaggedDocument(i,[str(j)]))
        j+=1
    return re        
```

```python
model = Doc2Vec(to_tagged(strs),dm=1,vector_size = 300,window = 4,epochs=100) #訓練doc2vec模型
```

```python
#用來預測檔案向量  直接把斷詞好後的句子丟進去就好 ex:['你好','嗎']
vec = model.infer_vector(strs[120],steps=20,alpha =0.001)  
```
